```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
it=pd.read_csv("./IT/data.csv")
it1=pd.read_csv("./IT/data (1).csv")
it2=pd.read_csv("./IT/data (2).csv")
it3=pd.read_csv("./IT/data (3).csv")
it4=pd.read_csv("./IT/data (4).csv")
it_oct=pd.read_csv("./data/oct_it.csv")
```

```python
metal=pd.read_csv("./METAL/data.csv")
metal1=pd.read_csv("./METAL/data (1).csv")
metal2=pd.read_csv("./METAL/data (2).csv")
metal3=pd.read_csv("./METAL/data (3).csv")
metal4=pd.read_csv("./METAL/data (4).csv")
metal_oct=pd.read_csv("./data/oct_metal.csv")
```
